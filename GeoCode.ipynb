```python
df = pd.read_csv('Mcdonald Location - Sheet1.csv')
```

```python
df
```

```python
lat = []
lng = []
for row in range(0, len(df)):
    address = df['Address'][row]
    suburb = df['Suburb'][row]
    state = df['State'][row]
    fulladdress = address + ',' + suburb + ',' + state
    g = geocoder.google(fulladdress)
    lat.append(g.lat)
    lng.append(g.lng)

df['latitude'] = lat
df['longtitude'] = lng
```

```python
df
```

```python
df.to_csv('Safeplacelocation.csv', index = False)
```
